```python
# Dependencies
import pandas as pd
```

```python
# Load CSV file as dataframe
diseasedata = pd.read_csv("Resources/U.S._Chronic_Disease_Indicators__CDI_.csv")
```

```python
# Preview the chronic diseases data
diseasedata.head()
```

```python
# Choose the relevant columns
chronic_df = diseasedata[["YearEnd", "LocationAbbr", "LocationDesc", "Topic", "Question", 
                          "DataValueUnit", "DataValueType", "DataValue", 
                          "Stratification1"]]
```

```python
# Preview the revised dataframe
chronic_df.head()
```

```python
# Select data for 2010, 2012, and 2014
Years = [2010, 2012, 2014]
chronic_df2 = chronic_df[chronic_df["YearEnd"].isin(Years)]
chronic_df2.head()
```

```python
# Choose data for diabetes
chronic_df3 = chronic_df2[chronic_df2["Topic"] == "Diabetes"]
chronic_df3.head()
```

```python
# Remove data for the US (we need state-wise data)
chronic_df4 = chronic_df3[chronic_df3["LocationAbbr"]!="US"]
chronic_df4.head()
```

```python
# Choose the data based on the following criteria:
# Crude Prevalence as the DataValueType
# Overall as the Stratification
chronic_df5 = chronic_df4.query("DataValueType == 'Crude Prevalence' & \
                                 Stratification1 == 'Overall'")
chronic_df5.head()
```

```python
# Choose data with Question as "Prevalence of diagnosed diabetes among adults aged >= 18 years"
chronic_df6 = chronic_df5[chronic_df5["Question"] == "Prevalence of diagnosed diabetes \
among adults aged >= 18 years"]

chronic_df6.head()
```

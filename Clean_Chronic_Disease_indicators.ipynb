```python
# Dependencies
import pandas as pd
```

```python
# Load CSV file as dataframe
diseasedata = pd.read_csv("Resources/U.S._Chronic_Disease_Indicators__CDI_.csv")
```

```python
# Preview the chronic diseases data
diseasedata.head()
```

```python
# Choose the relevant columns
clean_df = diseasedata[["YearEnd", "LocationAbbr", "LocationDesc", "Topic", "Question", "DataValueUnit", "DataValueType", "DataValue", "StratificationCategory1", "Stratification1"]]
```

```python
# Preview the revised dataframe
clean_df.head()
```

```python
# Select data for 2010, 2012, and 2014
Years = [2010, 2012, 2014]
clean_df2=clean_df[clean_df["YearEnd"].isin(Years)]
clean_df2.head()
```

```python
# Choose data for diabetes
clean_df2=clean_df2[clean_df2["Topic"]=="Diabetes"]
clean_df2.head()
```

```python
# Remove data for the US (we need state-wise data)
clean_df2=clean_df2[clean_df2["LocationAbbr"]!="US"]
clean_df2.head()
```
